```python
sheet1 = pandas.read_excel('RunTime_VS_Topology.xlsx', sheet_name='Лист1')
sheet2 = pandas.read_excel('RunTime_VS_Topology.xlsx', sheet_name='Изм.1')
```

```python
sheet1_2 = sheet1.loc[sheet1["Количество контрактов"] == 2]
sheet2_2 = sheet2.loc[sheet2["Количество контрактов"] == 2]

sheet1_3 = sheet1.loc[sheet1["Количество контрактов"] == 3]
sheet2_3 = sheet2.loc[sheet2["Количество контрактов"] == 3]

sheet1_2 = sheet1_2.astype({"Pyomo (сек)": 'float', "MIP (сек)": 'float'})
sheet1_3 = sheet1_3.astype({"Pyomo (сек)": 'float', "MIP (сек)": 'float'})

sheet2_2 = sheet2_2.astype({"Pyomo (сек)": 'float', "MIP (сек)": 'float'})
sheet2_3 = sheet2_3.astype({"Pyomo (сек)": 'float', "MIP (сек)": 'float'})
```

```python
sheet1_2.describe(), sheet2_2.describe()
```

```python
d1 = sheet1_2
title = "2 контракта"
x_label = "Число узлов"
d1 = d1.sort_values(by = [x_label])

d2 = sheet2_2
title = "2 контракта"
x_label = "Число узлов"
d2 = d2.sort_values(by = [x_label])
```

```python
f = plt.figure(figsize = (20, 10))
f.add_subplot(1, 2, 1)
plt.title("Было", fontsize = 30)
plt.plot(d1[x_label],d1["Pyomo (сек)"], label = "Pyomo")
plt.plot(d1[x_label],d1["MIP (сек)"], label = "MIP")
plt.xlabel(x_label)
plt.ylabel("Время выполнения")
plt.title(title)
plt.legend()

f.add_subplot(1, 2, 2)
plt.title("Стало", fontsize = 30)
plt.plot(d2[x_label],d2["Pyomo (сек)"], label = "Pyomo")
plt.plot(d2[x_label],d2["MIP (сек)"], label = "MIP")
plt.xlabel(x_label)
plt.ylabel("Время выполнения")
plt.title(title)
plt.legend()
plt.show()
```
